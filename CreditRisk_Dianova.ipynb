## Reduction of variables

### Feature Age

```python
def change_age(x):
    if x < 25:
        return 1
    elif x < 50:
        return 2
    else:
        return 3
```

```python
df2['Age'] = df2['Age'].apply(lambda x: change_age(x))
```

```python
print (df2['Age'].head())
```

### Feature Personal status and sex

```python
#`1`: `Male divorced/single`
#`2`: `Male married/widowed`
#`3`: `Female`

def change_sex(x):
    if x == 'male:single' or x == 'male:divorced/separated':
        return 1
    elif x == 'male:married/widowed':
        return 2
    else:
        return 3
```

```python
df2['Personal status and sex'] = df2['Personal status and sex'].apply(lambda x: change_sex(x))
```

```python
print (df2['Personal status and sex'].head())
```

# Model training (20-80)
Trenujeme na 20% dÃ¡t:

```python
from sklearn.model_selection import train_test_split
```

```python
X = df4_drop.drop(['Target'], axis=1)
Y = df4_drop.Target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)

print('Train and test size:', len(X_train), len(X_test))
print('Class balance: ', sum(Y)/len(Y))
```

```python
from sklearn import tree

# definition
model = tree.DecisionTreeClassifier()

# training
model = model.fit(X_train, Y_train)

# probabilities
Y_prob = model.predict_proba(X_test)

# predictions
Y_pred = model.predict(X_test)

# Accuracy on train set
model.score(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print('Accuracy: ', accuracy_score(Y_test, Y_pred)) 
print(classification_report(Y_test, Y_pred))

sns.heatmap(confusion_matrix(Y_test, Y_pred), fmt='g', annot=True, annot_kws={"size": 20})
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

**Accurancy** hovori o tom, na kolko percent mal model pravdu.
Vysledok je **True positive+True negative** = 116+35=151 z 200. Teda testovaci model bol na **75,5%** uspesny.

**False positive**: 24 z 200 = 12%. Tento model predikuje, ze 12% ziadatelov o uver splati svoj zavazok, no v skutonosti ho nesplatia.

**False negative**: A pri 12,5% (tj. 25 z 200) predikuje, ze svoj uver nesplatia, no v skutocnosti ho splatili.

# Tuning model (RECALL)

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'criterion': ['gini', 'entropy'],
              'max_depth': [2,3,5,10,12,15],
              'min_samples_split': [2,3,5,10],
              'min_samples_leaf': [1,2,3,5,10,12,15],
              'class_weight': [None, 'balanced']
             }

grid_search = GridSearchCV(
    tree.DecisionTreeClassifier(), 
    param_grid = param_grid, 
    scoring ='recall'
)

grid_search = grid_search.fit(X_train, Y_train)
best_params = grid_search.best_params_

print(best_params)
```

```python
# definition
model_tuned = tree.DecisionTreeClassifier(**best_params)

# training
model_tuned = model_tuned.fit(X_train, Y_train)

# probabilities
Y_prob_tuned = model_tuned.predict_proba(X_test)

# predictions
Y_pred_tuned = model_tuned.predict(X_test)

# Accuracy on train set
model_tuned.score(X_train, Y_train)
```

```python
print('Accuracy: ', accuracy_score(Y_test, Y_pred_tuned)) 
print(classification_report(Y_test, Y_pred_tuned))

sns.heatmap(confusion_matrix(Y_test, Y_pred_tuned), fmt='g', annot=True, annot_kws={"size": 20})
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

**Accurancy** sa sice znizila, ale model stale spravne predikoval spravanie 63% ziadatelov.

Nespravne identifikoval iba 6 zlyhanych uverov, co predstavuje **iba 3%-nu chybovost** vo false positive.

Inymi slovami, tento model predikuje, ze 3% ziadatelov splatia svoj uver, no v skutocnosti ho nesplatia.

Spravnu identifikaciu zlyhanych uverov sme ziskali na ukor zhorsenia **false negative** - poctu pripadov, kedy model predikoval, ze klient svoj uver nesplati no v skutocnosti ho splatil (68 z 200=34%). 

Z pohladu banky je velmi dolezite vediet spravne predikovat, kolko percent ziadatelov nesplati svoj uver, aby sa banka mohla spravne rozhodnut, v akej vyske si musi vytvorit svoje opravne polozky. 


Avsak dolezite je znizit chybovost modelu pri false negative, pretoze tym banka prichadza o zisk z poskytnutych uverov.

Preto sme namiesto metody **Recall** zvolili metodu **f1**.

# Hyperparameter Tuning Using Random Search

```python
# Load libraries
from scipy.stats import randint as sp_randint
from scipy.stats import uniform
from sklearn import linear_model, datasets
from sklearn.model_selection import RandomizedSearchCV
```

## Parameter Search for Random Forest

```python
param_dist = {
    "criterion": ["gini", "entropy"],
    "max_depth": sp_randint(1,16),
    "max_features": sp_randint(1, 11),
    "min_samples_split": sp_randint(2, 11),
    "min_samples_leaf": sp_randint(1,16),
    "n_estimators": [50]
}


random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(), 
    param_distributions=param_dist, 
#     n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1, 
    scoring="recall"
)

random_search = random_search.fit(X_train, Y_train)
best_params = random_search.best_params_

print(best_params)
```

```python
Y_prob = random_search.best_estimator_.predict_proba(X_test)
Y_pred = random_search.best_estimator_.predict(X_test)
```

```python
print('Accuracy: ', accuracy_score(Y_test, Y_pred)) 
print(classification_report(Y_test, Y_pred))

sns.heatmap(confusion_matrix(Y_test, Y_pred), fmt='g', annot=True, annot_kws={"size": 20})
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()

plot_roc_curve(Y_test, Y_prob, figsize=(6,6), curves=('each_class'))
plt.show()
```

**Najlepsie vysledky** sme dosiahli pomocou metody **Randomized Search CV**. Model ma najvyssie accurancy a aj najnizsiu chybovost pri False negative a False positive.

**Zaver**: Je teda na rozhodnuti danej banky, ci je jej cielom minimalizovat mnozstvo opravnych poloziek na nesplatene (defaultne) uvery alebo riskne defaultnost na ukor vyssieho zisku z vacsieho mnostva poskytnutych uverov.
